```python
# Instalasi library jika belum ada
!pip install -q tfx

import os
import tensorflow as tf
from tfx.orchestration.experimental.interactive.interactive_context import InteractiveContext
from tfx.components import CsvExampleGen, StatisticsGen, SchemaGen, ExampleValidator, Transform, Trainer, Evaluator, Pusher
from tfx.proto import trainer_pb2, pusher_pb2
```

```python
# --- GANTI DENGAN USERNAME ANDA ---
USERNAME = "raffihakim"
# -----------------------------------

PIPELINE_NAME = "student-score-pipeline-interactive"
PIPELINE_ROOT = os.path.join(f'{USERNAME}-pipeline')
DATA_ROOT = "data"
MODULE_FILE = os.path.join("pipeline", "module.py")
SERVING_MODEL_DIR = os.path.join(PIPELINE_ROOT, "serving_model", PIPELINE_NAME)

# Inisialisasi InteractiveContext
context = InteractiveContext(pipeline_root=PIPELINE_ROOT)
```

```python
# Mengambil data dari file CSV
example_gen = CsvExampleGen(input_base=DATA_ROOT)
context.run(example_gen)
```

```python
# Menghitung statistik data
statistics_gen = StatisticsGen(examples=example_gen.outputs['examples'])
context.run(statistics_gen)

# Menampilkan hasil statistik (opsional)
context.show(statistics_gen.outputs['statistics'])
```

```python
# Membuat skema data
schema_gen = SchemaGen(statistics=statistics_gen.outputs['statistics'], infer_feature_shape=True)
context.run(schema_gen)

# Menampilkan skema (opsional)
context.show(schema_gen.outputs['schema'])
```

```python
# Membuat skema data
schema_gen = SchemaGen(statistics=statistics_gen.outputs['statistics'], infer_feature_shape=True)
context.run(schema_gen)

# Menampilkan skema (opsional)
context.show(schema_gen.outputs['schema'])
```

```python
# Memvalidasi data
example_validator = ExampleValidator(
    statistics=statistics_gen.outputs['statistics'],
    schema=schema_gen.outputs['schema']
)
context.run(example_validator)

# Menampilkan anomali (opsional)
context.show(example_validator.outputs['anomalies'])
```

```python
# Melakukan feature engineering
transform = Transform(
    examples=example_gen.outputs['examples'],
    schema=schema_gen.outputs['schema'],
    module_file=MODULE_FILE
)
context.run(transform)
```

```python
# Melatih model
trainer = Trainer(
    module_file=MODULE_FILE,
    examples=transform.outputs['transformed_examples'],
    transform_graph=transform.outputs['transform_graph'],
    schema=schema_gen.outputs['schema'],
    train_args=trainer_pb2.TrainArgs(num_steps=5000),
    eval_args=trainer_pb2.EvalArgs(num_steps=1000)
)
context.run(trainer)
```

```python
# Mengevaluasi dan mendorong model
evaluator = Evaluator(
    examples=example_gen.outputs['examples'],
    model=trainer.outputs['model']
)
context.run(evaluator)

pusher = Pusher(
    model=trainer.outputs['model'],
    model_blessing=evaluator.outputs['blessing'],
    push_destination=pusher_pb2.PushDestination(
        filesystem=pusher_pb2.PushDestination.Filesystem(base_directory=SERVING_MODEL_DIR)
    )
)
context.run(pusher)
```

```python
import glob

# Cari path model terbaru yang di-push
latest_model_dir = max(glob.glob(os.path.join(SERVING_MODEL_DIR, '*/')), key=os.path.getmtime)
print(f"Loading model from: {latest_model_dir}")
loaded_model = tf.saved_model.load(latest_model_dir)

def predict_performance(model, data):
    infer = model.signatures['serving_default']
    input_tensors = {key: tf.constant([[value]]) for key, value in data.items()}
    predictions = infer(**input_tensors)
    return predictions['output_0'].numpy()[0][0]

# Skenario tes
siswa_positif = {
    'Hours_Studied': 9.0, 'Attendance': 95.0, 'Parental_Involvement': 4.0, 
    'Sleep_Hours': 8.0, 'Previous_Scores': 90.0, 'Motivation_Level': 9.0, 
    'Family_Income': 80000.0, 'Teacher_Quality': 5.0, 'Peer_Influence': 4.0, 
    'Physical_Activity': 4.0, 'Distance_from_Home': 5.0,
    'Access_to_Resources': 'yes', 'Extracurricular_Activities': 'yes',
    'Internet_Access': 'yes', 'Tutoring_Sessions': 'yes', 'School_Type': 'private',
    'Learning_Disabilities': 'no', 'Parental_Education_Level': 'graduate', 'Gender': 'female'
}

prediksi = predict_performance(loaded_model, siswa_positif)
print(f"\nPrediksi Skor untuk Siswa dengan Atribut Positif: {prediksi:.2f}")
```
